##### Borough containing the word Toronto

##### Map of Toronto, Canada

```python
import folium
```

```python
latitude = 43.6532
longitude = -79.3832 

TorontoMap = folium.Map(location=[latitude, longitude], zoom_start=11)
TorontoMap
```

Adding marker onto the Map

```python
for lat, lng, label in zip(Toronto_df['Latitude'], Toronto_df['Longitude'], Toronto_df['Postcode']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(TorontoMap)  

TorontoMap
```
